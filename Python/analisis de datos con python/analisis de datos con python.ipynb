```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], ignore_index= True)
```

```python
pd.concat([df1,df2], axis = 1)
```

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
```

```python
izq
```

```python
der
```

```python
izq.merge(der, on='key')
```

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2','k3'],
 'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
```

```python
izq.merge(der, left_on = 'key', right_on='key_2')
```

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2',np.nan],
 'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
```

```python
izq.merge(der, left_on = 'key', right_on='key_2') # sin el how, se hace un innerjoin,trayendo unicamente los valores que coinciden
```

```python
izq.merge(der, left_on = 'key', right_on='key_2', how='left') # si le doy inner en how, se respeta el anterior codigo
```

```python
izq.merge(der, left_on = 'key', right_on='key_2', how='right') # al darle right, se queda unicamente lo que cruzo en la derecha
```

join, lo hace por index y no por columnas

```python
izq = pd.DataFrame({'A': ['A0','A1','A2'],
  'B':['B0','B1','B2']},
  index=['k0','k1','k2'])

der =pd.DataFrame({'C': ['C0','C1','C2'],
  'D':['D0','D1','D2']},
  index=['k0','k2','k3']) 
```

```python
izq.join(der)
```

```python
izq.join(der, how='inner')
```

```python
izq.join(der, how='left')
```

```python
izq.join(der)
```

```python
izq.join(der, how = 'outer')
```
