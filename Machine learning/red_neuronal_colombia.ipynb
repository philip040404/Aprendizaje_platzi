```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Inicializar la RNA
classifier = Sequential()

# Aadir las capas de entrada y primera capa oculta
classifier.add(Dense(units = 43, kernel_initializer = "normal",  
                     activation = "relu", input_dim = 87))

# Aadir la segunda capa oculta
classifier.add(Dense(units = 43, kernel_initializer = "normal",  activation = "relu"))

classifier.add(Dense(units = 43, kernel_initializer = "normal",  activation = "relu"))

classifier.add(Dense(units = 43, kernel_initializer = "normal",  activation = "relu"))

classifier.add(Dense(units = 43, kernel_initializer = "normal",  activation = "relu"))

classifier.add(Dense(units = 43, kernel_initializer = "normal",  activation = "relu"))


# Aadir la capa de salida
classifier.add(Dense(1))

# Compilar la RNA
classifier.compile(optimizer = "adam", loss = "mean_squared_error", metrics = ['MeanSquaredError','AUC'])

# Ajustamos la RNA al Conjunto de Entrenamiento
classifier.fit(X_train, y_train,  batch_size = 10, epochs = 500)


# Parte 3 - Evaluar el modelo y calcular predicciones finales

# Prediccin de los resultados con el Conjunto de Testing
```
