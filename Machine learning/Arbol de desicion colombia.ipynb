```python
# Regresión Bosques Aleatorios

# Cómo importar las librerías
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_excel('./Material/Datasets/Modelo 1.xlsx',sheet_name='Sheet3')
```

```python
dummy_variables1 = pd.get_dummies(dataset['Plant'])
dummy_variables2 = pd.get_dummies(dataset['Route'])
dummy_variables3 = pd.get_dummies(dataset['Sold To Name'])
dummy_variables4 = pd.get_dummies(dataset['BU'])
```

```python
dummy_variables4
```

```python
dataset = pd.concat([dataset, dummy_variables1,dummy_variables2,dummy_variables3,dummy_variables4], axis=1)
```

```python
dataset['time']=dataset['tiempo de entrega']
```

```python
dataset.drop(['Plant','Sold To Name','Route','tiempo de entrega','BU'],inplace=True,axis=1)
```

```python
dataset.shape
```

```python
X = dataset.iloc[:, 1:69].values
y = dataset.iloc[:, 69].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X
```

```python
dataset
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
regression = RandomForestRegressor(n_estimators = 2000, random_state=42)
regression.fit(X_train,y_train)
```

```python
y_train
```

```python
y_predi = regression.predict(X_test)
```

```python
y_predi
```

```python
from sklearn.metrics import r2_score
# obtener predicciones para el conjunto de prueba
y_pred = regression.predict(X_test)
# calcular R2
r2 = r2_score(y_test, y_pred)
print('R2:', r2)
```

```python
from sklearn.metrics import mean_squared_error

# obtener predicciones para el conjunto de prueba
y_pred = regression.predict(X_test)

# calcular MSE
mse = mean_squared_error(y_test, y_pred)

print('MSE:', mse)
```
