```python
from openpyxl import Workbook

# Crear un nuevo libro de Excel
wb = Workbook()
ws = wb.active

# Datos de ejemplo
datos = [
    ['valor', 'otra_columna'],  # Encabezados de columna
    ['otoño', 'dato1'],
    ['verano', 'dato2'],
    ['invierno', 'dato3'],
    ['verano', 'dato4'],
]

# Agregar datos al libro de Excel
for fila in datos:
    ws.append(fila)

# Filtrar las filas en el libro de Excel
filas_filtradas = [fila for fila in ws.iter_rows(values_only=True) if fila[0] == 'verano']

# Crear un nuevo libro para las filas filtradas
wb_filtrado = Workbook()
ws_filtrado = wb_filtrado.active

# Agregar filas filtradas al nuevo libro
for fila in filas_filtradas:
    ws_filtrado.append(fila)

# Guardar el libro de Excel filtrado
wb_filtrado.save('archivo_filtrado.xlsx')
```

```python
from openpyxl import Workbook

# Crear un nuevo libro de Excel
wb = Workbook()
ws = wb.active

# Datos de ejemplo con diferentes variaciones de 'verano'
datos = [
    ['valor', 'otra_columna'],  # Encabezados de columna
    ['otoño', 'dato1'],
    ['Verano', 'dato2'],
    ['VErano', 'dato3'],
    ['VERAno', 'dato4'],
    ['vérano', 'dato5'],
]

# Agregar datos al libro de Excel
for fila in datos:
    ws.append(fila)

# Normalizar los datos en la columna 'A' a minúsculas antes de aplicar el filtro
for fila in ws.iter_rows(min_row=2, max_row=ws.max_row, min_col=1, max_col=1):
    for celda in fila:
        celda.value = celda.value.lower()

# Aplicar filtro en la columna 'A' para mostrar solo las filas que contienen 'verano'
ws.auto_filter.ref = ws.dimensions
ws.auto_filter.add_filter_column(0, ['verano'])

# Guardar el libro de Excel con el filtro aplicado
wb.save('archivo_con_filtro_veranodd.xlsx')
```

```python
from openpyxl import Workbook

# Crear un nuevo libro de Excel
wb = Workbook()
ws = wb.active

# Datos de ejemplo con diferentes variaciones de 'verano'
datos = [
    ['valor', 'otra_columna'],  # Encabezados de columna
    ['otoño', 'dato1'],
    ['Verano', 'dato2'],
    ['VErano', 'dato3'],
    ['VERAno', 'dato4'],
    ['vérano', 'dato5'],
]

# Agregar datos al libro de Excel
for fila in datos:
    ws.append(fila)

# Normalizar los datos en la columna 'A' a minúsculas antes de aplicar el filtro
for fila in ws.iter_rows(min_row=2, max_row=ws.max_row, min_col=1, max_col=1):
    for celda in fila:
        celda.value = celda.value.lower()

# Ocultar las filas que no contienen 'verano'
for fila in range(2, ws.max_row + 1):
    if ws.cell(row=fila, column=1).value != 'verano':
        ws.row_dimensions[fila].hidden = True

# Guardar el libro de Excel con las filas ocultas
wb.save('archivo_con_filtro_veranossss.xlsx')
```
